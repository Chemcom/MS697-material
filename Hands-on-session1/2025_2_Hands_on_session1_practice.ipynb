# **Advanced examples (other materials datasets)**

The matminer.datasets module provides benchmark datasets that are pre-processed and ready to use for machine learning research in materials science. Unlike MPDataRetrieval or AFLOWDataRetrieval, these do not require an external API key — you can load them directly as a pandas DataFrame.

Here are some computational and experimental datasets you can load:

brgoch_superhard_training
: 2,574 materials used to train regressors for shear and bulk modulus prediction.

castelli_perovskites
: 18,928 perovskites generated via ABX combinatorics. Includes GLLB-SC band gap, PBE structure, absolute band edge positions, and formation enthalpies.

citrine_thermal_conductivity
: Experimental thermal conductivity of 872 compounds from the Citrine database. Includes various temperatures (295 at room temperature).

dielectric_constant
: 1,056 structures with dielectric properties calculated using DFPT-PBE.

elastic_tensor_2015
: 1,181 structures with elastic properties computed by DFT-PBE.

expt_formation_enthalpy
: 1,276 entries of experimental formation enthalpies (mostly binaries), with matching MPIDs/OQMDIDs where available.

expt_formation_enthalpy_kingsbury
: Cleaned dataset of experimental standard formation enthalpies (Kim et al., NIST JANAF, Kubaschewski). Mapped to MP IDs where possible.

expt_gap
: Experimental band gaps of 6,354 inorganic semiconductors.

expt_gap_kingsbury
: Same as matbench_expt_gap, but with MP-ID mapping.

matbench_* (e.g., matbench_expt_gap, matbench_perovskites, matbench_log_kvrh, etc.)
: Benchmark datasets for ML tasks from the Matbench benchmark suite (band gaps, moduli, metallicity, phonons, steels, etc.).

steel_strength
: 312 steels with experimental yield and tensile strength.

superconductivity2018
: about 16,000 superconductivity records (Tc) from Stanev et al. (~300 Tc=0).

ucsb_thermoelectrics
: ~1,100 experimental thermoelectric materials (UCSB database, from 108 publications).

```python
import plotly.express as px
from matminer.datasets import load_dataset

df = load_dataset('elastic_tensor_2015')

fig = px.scatter(
    df,
    x="G_VRH",
    y="K_VRH",
    color="poisson_ratio",
    hover_name="material_id",
    color_continuous_scale="Picnic",
    range_x=[0, 300],
    labels={
        "G_VRH": "Shear Modulus (GPa)",
        "K_VRH": "Bulk Modulus (GPa)"
    }
)
fig.update_traces(marker=dict(line=dict(width=1, color="black")))
fig.show()
```
